```python
def histogram(image, *args):
    #checks wheter args is empty
    if args:
        max_intensity = args[0]
    else:
        max_intensity = npy.amax(image)
    
    hist = npy.zeros(max_intensity + 1).astype(int)
    for i in range(max_intensity + 1):
        #size of the returned array
        pixel_count = npy.where(image == i)[0].shape[0]
        hist[i] = pixel_count
    if args:
        return hist
    else:
        return (hist, max_intensity + 1)
```

```python
hist_img_1, img_1_imax = histogram(img_1)

plt.figure(figsize=(10, 10))
#1 row 2 cols 1st position
plt.subplot(121)
plt.imshow(img_1, cmap="gray", vmin=0, vmax=255)
plt.axis('off')
plt.subplot(122)
plt.bar(range(img_1_imax), hist_img_1)
plt.xlabel("Intensities")
plt.ylabel("Frequency")
```
